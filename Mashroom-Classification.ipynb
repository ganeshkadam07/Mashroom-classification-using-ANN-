```python
# Assign Dependent and Independent Variable
x=df.drop('class_e',axis=1).values
y=df['class_e'].values
```

```python
x.shape
```

```python
# Split Data into training and Testing set
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Deep Learning ANN Classifier

```python
# Built a ANN Classifier model 
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
res=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=32)
```

```python
# Training Accuracy
model.evaluate(x_train,y_train)
```

```python
# Testing Accuracy
model.evaluate(x_test,y_test)
```

```python
# Accuracy Plot
plt.plot(res.history['accuracy'],label="Accuracy")
plt.plot(res.history['val_accuracy'],label="Val_Accuracy")
```

```python
# graph of Loss
plt.plot(res.history['loss'],label="Loss")
plt.plot(res.history['val_loss'],label="Val_loss")
```
